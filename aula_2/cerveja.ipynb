```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("../data/dados_cerveja_nota.xlsx")
```

```python
df.head()
```

```python
plt.plot(df["cerveja"],df["nota"],"o")
plt.grid(True)
plt.xlim(0,11)
plt.ylim(0,11)
plt.xlabel("Cerveja")
plt.ylabel("Nota")
plt.title("Relação Nota x Cerveja")
plt.show()
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[["cerveja"]],df[["nota"]])

print("Intercepto :", reg.intercept_)
print("Coeficientes :", reg.coef_)
```

```python
X = df[["cerveja"]]
y_estimado = reg.predict(X)
```

```python
plt.plot(df["cerveja"],df["nota"],"o")
plt.plot(X,y_estimado,"-")
plt.grid(True)
plt.xlim(0,11)
plt.ylim(0,11)
plt.xlabel("Cerveja")
plt.ylabel("Nota")
plt.title("Relação Nota x Cerveja")
plt.show()
```

```python
from sklearn import tree

arvore = tree.DecisionTreeRegressor(max_depth=2)
arvore.fit(df[["cerveja"]],df[["nota"]])
```

```python
y_estimado_arvore = arvore.predict(X)
```

```python
plt.plot(df["cerveja"],df["nota"],"o")
plt.plot(X,y_estimado,"-")
plt.plot(X,y_estimado_arvore,"-")
plt.grid(True)
plt.xlim(0,11)
plt.ylim(0,11)
plt.xlabel("Cerveja")
plt.ylabel("Nota")
plt.title("Relação Nota x Cerveja")
plt.legend(["Pontos","Regressão","Arvore"])
plt.show()
```
